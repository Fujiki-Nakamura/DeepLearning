```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import random
```

```python
def plot_horizontally(data, sample_i, ncol=10, figsize=(16, 16)):
    plt.figure(figsize=figsize)
    for i in range(10):
        plt.subplot(1, ncol, i + 1)
        plt.imshow(data[sample_i][i], 'gray')
        plt.axis('off')
    plt.show()
```

```python
exp = '/20190915050748/epoch96/'
inpts = np.load('../../logs/{}/inpts.npy'.format(exp), allow_pickle=True)
preds = np.load('../../logs/{}/preds.npy'.format(exp), allow_pickle=True)
trues = np.load('../../logs/{}/trues.npy'.format(exp), allow_pickle=True)

exp = '/20190915050748/epoch35/'
inpts_best = np.load('../../logs/{}/inpts.npy'.format(exp), allow_pickle=True)
preds_best = np.load('../../logs/{}/preds.npy'.format(exp), allow_pickle=True)
trues_best = np.load('../../logs/{}/trues.npy'.format(exp), allow_pickle=True)
```

```python
sample_i = random.randint(0, 1000)
plot_horizontally(inpts, sample_i)
plot_horizontally(trues, sample_i)
plot_horizontally(preds, sample_i)

plot_horizontally(preds_best, sample_i)
```
